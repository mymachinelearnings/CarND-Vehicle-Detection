### Histogram of Oriented Gradients
Training a classifier with the exact colors or shapes of the image would not work very well as the size and shapes as well as colors of the vehicles change often depending on the point of view of hte camera as well as shadows/lightness. Hence we use a technique called Histogram of Oriented Gradients(HOG)

In HOG, we sample the entire image in to multiple smaller cells, and we calculate the gradient of each pixel in the cell. Now a circular histogram (of a specific number of bins) is taken out of the magnitude of gradients of each pixel in the cell and is represented in the cell instead of the actual pixels. This gives us a flexibility to memorize the shape of the image and at the same time giving flexibility for the change of colors or orientation

sklearn python library gives predefined function to calculate HOG features. This takes the following arguments
- image : Input Image
- Orientations : Number of orientations(bins) in the circular histogram
- Pixels per cell : width and height of the individual cell
- Cell per block : number of cells to be considered while calculating the gradient magnitude

### Combining features
While HOG is a pretty good technique to evaluate the features of an image, we can combine it with other methods like 

The classifier will be more robust if we consider various features of the data that is useful. 

In this case, we are combining the following features
- Spatial Binning of the Image
- Histogram of Colors
- HOG Features

This process enables to hold more information of the same image compared to only considering a single feature
After all the features are extracted, they are stacked to a 1D vector

Note : It is important to note that as you keep on combining different features, the vector size increases and hence the computatoinal ask. Hence stick on to a optimum level of combining features.

Also note that whatever feature extraction is done for training the classifier, the SAME scheme of feature extraction should be done for the test set (and the actual video feed) as well

Once all these are evaluated, all htese can be combined over a 1D vector
#### Normalization
It is VERY important to normalize when you are combining different features of an image, otherwise you may end up in a situation where some values override the other since each feature extraction follows its own scale (for ex, gradient magnitude scale is different from color space(0-255))

Normalization can be achieved by sklearn StandardScaler function

### Training a Classifier
We are using Linear SVM classifier for this project which is a good choice for Logistic Regression case where the output is either Yes or No (or basically 2 values)

Python sklearn library provides readily available function `LinearSVC()` to define a classifier. We can then use `fit(features, label)` to train the classifier. The accuracy of the classifier can be evaluated using `score` function, and prediction can be made using `predict` function

Note : It is a good practice to store the classifier and other parameters into a pickled dataset so that it will be later used. Training a classifier is computationally intensive, so we don't want to retrain it every time we need the classifier object(when the server is restarted, or in a different python file)

### Searching Image for Cars
I've used two approaches to search the cars in an image

**Approach 1**
This is the basic approach which is computationally intensive.
First, we determine the window size (in pixels) and slide it through the entire span of the image. You can select overlap between two consequtive images so that you capture more information, and not to miss the information at the edges.

We know that most most of the top half of the image consists of trees, sky and others which is not our region of interest. Hence we can filter these by specifying the range of y while evaluating the windows.

Now, each of the window identified is passed to a helper function to evaluate the features which are then passed to the classifier to see if this contains a car or not. Based on the prediciton of the classifier, we save the window to a list which will be finally used to identify the bounding boxes

Problem with Approach 1 :
It is obvious that we are calculating the features (HOG, Color Histogram & Spatial binning) for every window seperately. And due to overlap, we are calculating the same again and again at some places. To avoid this, we'll follow approach 2

**Approach 2**
In approach 2, we calculate the spatial features at once for whole image. The image will be then splitted into much smaller cells. Window of interest could be to the level of each cell, or a combination of cells. Based on the window selection, we determine the corresponding features from the full feature extraction and use that to predict.

#### Scaling
As you know, the size of the car changes in the image depending on its distance from the camera. Hence sticking to a fixed window width might not evaluate cars of different sizes, at different positions. Hence it is a good practice to scale the windows to a lower and higher values so that cars of all sizes(relative) can be identified. This is done by using a scaling factor and changing the window size based on the scaling factor. The classification logic will still remain the same.

### Dealing with False Positives
Either in Approach 1 or 2 of finding cars in the image, it is apparent that we come across false positives. One simple way to deal with it is with the help of heatmaps.

Heatmap is nothing but a series of zeros, and everytime a bounding box is identified, we +1 on it so that it gets highlighted. More identifications at a particular place means more value to it which results in a brighter portion in the heat map.

Once the heat map is identified, we'll use a minumum threshold to remove false positives. Obviously false positives are unwanted data which won't be consistent. Hence by checking the number of occurances of such false positives(bounding boxes), we can filter them out so that we are left with only those that appear over a certain number(threshold) of times

This results in a valid classification in the heatmap. We can then create a bounding box over the entire range of the highlights in the heatmap and label them using the `scipy.ndimage.measurements`'s `Label` function.

This results in a complete single bounding box per vehicle
